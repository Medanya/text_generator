```python
from glob import glob
import re 
import bleach
from bs4 import BeautifulSoup

def has_cyrillic(text):
    return bool(re.search('[а-яА-Я]', text))

def cleanhtml(raw_html):
    cleanr = re.compile('<.*?>')
    nourl = re.compile(r'\w+:\/{2}[\d\w-]+(\.[\d\w-]+)*(?:(?:\/[^\s/]*))*')
    cleantext = re.sub(cleanr, '', raw_html)
    cleantext = re.sub(nourl, '', cleantext)
    return cleantext.rstrip()
```

```python
names = {"Андрей Максимов", 'Игорь Сошилов', "Александр Смирнов", "Андрей Решетников", "Аня Журавлева"} 
def filter_html(path):
    data = open(path, 'r').readlines()
    data = list(map(cleanhtml, data))
    data = list(filter(lambda x: len(x) > 5, data))
    data = list(filter(lambda x: x[0].isalpha(), data))
    data = list(filter(lambda x: has_cyrillic(x[0]), data))
    data = list(filter(lambda x: x not in names, data))
    return data
```

```python
files = glob("./Telegram Desktop/gluxar/*.html") + glob("./Telegram Desktop/lesha-me/*.html")
```

```python
data = list()

for f in files:
    data.extend(filter_html(f))
```

```python
print(len(data))
```

```python
with open("./Telegram Desktop/talks.txt", 'w') as f:
    for l in data:
        f.write(l + "\n")
```
